checking_status: Status of the existing checking account. It typically represents the balance or status of a customer's checking account, indicating financial stability.


duration: The duration of the credit in months. This indicates how long the credit is expected to be paid back.


credit_history: Describes the credit history of the individual. It may include values such as 'no credits/all paid', 'existing paid', 'critical/other existing credit', which indicate the individual's past behavior in managing credit.


purpose: The purpose of the credit or loan. Common purposes might include 'new car', 'used car', 'furniture', 'education', etc.


credit_amount: The amount of credit or loan requested by the individual. This is the principal amount of the loan.

installment_commitment: The percentage of disposable income that will go towards the installment payments for the credit.


personal_status: The personal and marital status of the individual, as discussed earlier (e.g., 'male single', 'female div/dep/mar').


other_parties: Indicates if there are other parties involved in the loan, such as a guarantor or co-applicant.


residence_since: The number of years the individual has lived at their current residence. This might indicate stability.


property_magnitude: The value or type of property owned by the individual, possibly categorized into different ranges or types like 'real estate', 'building society savings agreement/life insurance', 'car', etc.


age: The age of the individual in years.


other_payment_plans: Indicates if the individual has other payment plans, such as from a bank or a store.


housing: The individual's housing situation, like 'own', 'rent', or 'for free'.


existing_credits: The number of existing credits at the bank, which indicates how many other loans the individual currently has.


job: The type of job the individual has, which may be categorized by profession or employment status (e.g., 'skilled', 'unskilled', 'management').


num_dependents: The number of people dependent on the individual, typically children or other family members.


own_telephone: Indicates whether the individual has a telephone, which might be a factor in assessing their stability or reliability.


foreign_worker: Indicates whether the individual is a foreign worker, which might be used to assess risk.


class: The target variable indicating the classification or risk category of the credit, such as 'good' or 'bad' credit risk.

## Understanding Data Types and Unique Values of each column

```python
df.info()
```

```python
for column in df.columns:
    print("Column name: ",column)
    print("Data Type: ",df[column].dtype)
    print("Unique Values: ",sorted(df[column].unique()))
    print("\n")
```

1) The Elements in each column are consistent with their respective column's datatype

2) The column 'credit_history' contains some values eqaul to 'no credits/all paid' and some values equal to 'all paid'. So Values equal to 'no credits/all paid' needs to be replaced to 'no credits' in 'credit_history' column to avoid confusion and maintain uniqueness of each sample.

3) The column 'personal status' can be divided into two columns: one for gender and other for marital status.

```python
df.replace({
    'credit_history' : {
        'no credits/all paid' : 'no credits'
    }
},
inplace = True)
```

```python
# Mapping gender based on presence of 'male' or 'female'
df['gender'] = df['personal_status'].apply(lambda x: 'female' if 'female' in x else 'male')

# Mapping marital status based on keywords
def extract_marital_status(status):
    if 'single' in status:
        return 'single'
    elif 'mar/wid' in status:
        return 'married/widowed'
    elif 'div/sep' in status:
        return 'divorced/separated'
    elif 'div/dep/mar' in status:
        return 'divorced/dependent/married'
    return 'unknown'

df['marital_status'] = df['personal_status'].apply(extract_marital_status)

df.drop('personal_status', axis=1, inplace=True)
```

### Bivariate Analysis between categorical attributes and target class

Since our target class is highly imbalanced, we use proportion for analysis instead of raw counts

```python
categorical_columns = [x for x in df.columns if x not in numerical_columns]
categorical_columns
```

```python
ct = pd.crosstab(df['class'],df['checking_status'], normalize = 'index').reindex(columns = ['no checking', '<0', '0<=X<200', '>=200'])

print(ct)

# Proportion Bar Plot
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
ct.plot(kind='bar', ax=plt.gca(), color=sns.color_palette('Set2'))
plt.title(f'Proportion of checking_status by Class')
plt.xlabel('class')
plt.ylabel('Proportion')
plt.legend(title='checking_status')

# Stacked Bar Plot
plt.subplot(1, 2, 2)
ct.plot(kind='bar', stacked=True, ax=plt.gca(), color=sns.color_palette('Set2'))
plt.title(f'Stacked Bar Chart of checking_status by Class')
plt.xlabel('class')
plt.ylabel('Proportion')

plt.tight_layout()
plt.show()
```

Conclusion:
1) Majority people with good credit have no checking status whereas majority people with bad credit have less than 0 in their checking account.

2) Very few people with more than 200 in their checking have bad credit

3) In bad credit population, there are more people with less than 0 in checking account than those with 0 to 200 in contrast to good credit population.

## Feature Importance Measuring Techniques

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

model = RandomForestClassifier(random_state=42)

# Fit the model
model.fit(df[numerical_columns], df['class'])

importances = model.feature_importances_
feature_importance_df = pd.DataFrame({'Feature': numerical_columns, 'Importance': importances})

feature_importance_df.sort_values(by='Importance', ascending=False, inplace=True)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance')
plt.show()
```

```python
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder

# Convert categorical features and target to numerical
X_encoded = df[categorical_columns].drop('class',axis=1).apply(LabelEncoder().fit_transform)
y_encoded = LabelEncoder().fit_transform(df['class'])

# Perform chi-square test
chi2_values, p_values = chi2(X_encoded, y_encoded)

# Create a DataFrame to view the results
importance_df = pd.DataFrame({'Feature': X_encoded.columns, 'Chi2': chi2_values, 'p-value': p_values})
importance_df = importance_df.sort_values(by='Chi2', ascending=False)
importance_df
```

```python
from sklearn.feature_selection import f_classif

# Assume X_encoded is one-hot encoded
f_values, p_values = f_classif(X_encoded, y_encoded)

# Create a DataFrame to view the results
importance_df = pd.DataFrame({'Feature': X_encoded.columns, 'F-Value': f_values, 'p-value': p_values})
importance_df = importance_df.sort_values(by='F-Value', ascending=False)
importance_df
```

```python
from scipy.stats import pointbiserialr

# Calculate Point Biserial Correlation for each feature
correlations = []
for column in X_encoded.columns:
    corr, _ = pointbiserialr(y_encoded, X_encoded[column])
    correlations.append((column, corr))

# Create a DataFrame to view the results
correlation_df = pd.DataFrame(correlations, columns=['Feature', 'Point Biserial Correlation'])
correlation_df = correlation_df.sort_values(by='Point Biserial Correlation', ascending=False)
correlation_df
```

```python
from sklearn.feature_selection import mutual_info_classif

# Calculate Mutual Information
mi_scores = mutual_info_classif(X_encoded, y_encoded)

# Create a DataFrame to view the results
mi_df = pd.DataFrame({'Feature': X_encoded.columns, 'Mutual Information': mi_scores})
mi_df = mi_df.sort_values(by='Mutual Information', ascending=False)
mi_df
```

```python
from sklearn.ensemble import RandomForestClassifier

# Fit the model
model = RandomForestClassifier()
model.fit(X_encoded, y_encoded)

# Get feature importance
importances = model.feature_importances_

# Create a DataFrame to view the results
importance_df = pd.DataFrame({'Feature': X_encoded.columns, 'Importance': importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
importance_df
```
